```python
# Import libraries we need

import json # For reading JSON into a dictionary
import os # Operating system utility functions, e.g. listing all JSON files in our archive

# Define constants

ARCHIVE_DIR = '../archive/' # where the archived files are
print(ARCHIVE_DIR)
```

```python
# Open the most recent archived file in, and load the JSON as a Python object (in this case, a list)

archive_filename = os.listdir(ARCHIVE_DIR)[0] # List of all the archive files, then take the first one
courses = json.load(open(ARCHIVE_DIR + archive_filename))

print(archive_filename)
print(len(courses))
```

```python
all_courses = [] 

for course in courses:
    
     for section in course['course']['classes']['class']:
            
        #GS/CC LitHum
        if 'EURPN LIT-PHILOS MASTERPIECS I' in section['title']:
            all_courses.append({
                'instructor_uni': section['instructors'][0]['uni'],
                'title': section['title'],
                'enrollment_count': section['enrollment']['count'],
                'subject': course['course']['subject']['long_name'],
               # 'course_id': course['course']['course_identifier']
                })
        #UW 
        if 'UNIVERSITY WRITING' in section['title']: 
            all_courses.append({
                'instructor_uni': section['instructors'][0]['uni'],
                'title': section['title'],
                'enrollment_count': section['enrollment']['count'],
                'subject': course['course']['subject']['long_name']
                })
            
        #Contemp. Civilizations 
        if course['course']['subject']['long_name'] == 'Contemporary Civilization':
            all_courses.append({
              'instructor_uni': section['instructors'][0]['uni'],
              'title': section['title'],
              'enrollment_count': section['enrollment']['count'],
              'subject': course['course']['subject']['long_name']
            })
        
   
    
#there are 80 UW courses which makes sense bc 26 GS and 54 CC 
            
print(all_courses)
with open('all_courses.json', 'w') as f:
   json.dump(all_courses, f)


    
```
